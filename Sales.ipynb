```python
#import the necessary library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load the dataset
store = pd.read_csv('/content/store_sales.csv')
store.head()
```

```python
#check for the numbers of rows and columns
store.shape
```

```python
#check for missing values
store.isnull().sum()
```

```python
#drop the missing values
store = store.dropna()
```

```python
#check for missing values again to be sure it has be dropped
store.isna().sum()
```

```python
#get information about the dataset
store.info()
```

```python
# check for duplicates
store.duplicated().sum()
```

```python
#plot the relative frequency
plt.figure(figsize=(20, 5))

plt.subplot(1, 4, 1)
sns.set_style("darkgrid")
store['Category'].value_counts(normalize=True).plot(kind="bar", color='#1a335c')
plt.xlabel("Category", **{'weight':'bold', 'color':'#2a2e33'})
plt.ylabel("Relative Frequency", **{'weight':'bold', 'color':'#2a2e33'})
plt.title('Relative Frequency of Category', **{'weight':'bold', 'color':'#2a2e33'})

plt.subplot(1, 4, 2)
sns.set_style("darkgrid")
store['Region'].value_counts(normalize=True).plot(kind="bar", color='#112410')
plt.xlabel("Region", **{'weight':'bold', 'color':'#2a2e33'})
plt.ylabel("Relative Frequency", **{'weight':'bold', 'color':'#2a2e33'})
plt.title('Relative Frequency of Region', **{'weight':'bold', 'color':'#2a2e33'})


plt.subplot(1, 4, 3)
sns.set_style("darkgrid")
store['Ship Mode'].value_counts(normalize=True).plot(kind="bar", color='#5d2563')
plt.xlabel("Ship Mode", **{'weight':'bold', 'color':'#2a2e33'})
plt.ylabel("Relative Frequency", **{'weight':'bold', 'color':'#2a2e33'})
plt.title('Relative Frequency of Ship Mode', **{'weight':'bold', 'color':'#2a2e33'})


plt.subplot(1, 4, 4)
sns.set_style("darkgrid")
store['Segment'].value_counts(normalize=True).plot(kind="bar", color='#633925')
plt.xlabel("Segment", **{'weight':'bold', 'color':'#2a2e33'})
plt.ylabel("Relative Frequency", **{'weight':'bold', 'color':'#2a2e33'})
plt.title('Relative Frequency of Segment', **{'weight':'bold', 'color':'#2a2e33'})




plt.show()

```

```python
store.groupby(['Category','Sub-Category'])['Sales'].agg(["sum","count"]).round()
```

```python
plt.figure(figsize=(12, 8))

store['Sub-Category'].value_counts(normalize=True).plot(kind="bar", color='#d19c0a')
plt.xlabel("Sub-Category", **{'weight':'bold', 'color':'#2a2e33'})
plt.ylabel("Relative Frequency", **{'weight':'bold', 'color':'#2a2e33'})

plt.title('Relative Frequency of Sub_Category', **{'weight':'bold', 'color':'#2a2e33'})
plt.xticks(rotation=90)
plt.show()
```

```python
# plot a pie
plt.figure(figsize=(21, 5))

plt.subplot(1, 4, 1)
store.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette("dark:salmon_r"), textprops={'weight':'bold', 'color':'#2a2e33'})
plt.title('Region',  **{'weight':'bold', 'color':'#2a2e33'})

plt.subplot(1, 4, 2)
store.groupby('Category')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette("dark:salmon_r"), textprops={'weight':'bold', 'color':'#2a2e33'})
plt.title('Category',  **{'weight':'bold', 'color':'#2a2e33'})

plt.subplot(1, 4, 3)
store.groupby('Ship Mode')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette("dark:salmon_r"), textprops={'weight':'bold', 'color':'#2a2e33'})
plt.title('Ship Mode',  **{'weight':'bold', 'color':'#2a2e33'})

plt.subplot(1, 4, 4)
store.groupby('Segment')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette("dark:salmon_r"), textprops={'weight':'bold', 'color':'#2a2e33'})
plt.title('Segment',  **{'weight':'bold', 'color':'#2a2e33'})

plt.show()
```



## Regions
- The west region has the highest total sum of sales accounting for 31.5%, while the south region has only 17.5%, which is the lowest.

## Categories
- For the category, technology has the highest total sum of sales with 36.7% of the total sale, followed by furniture with 32.1%, and lastly office supplies with 31.2%.

## Ship Mode
- The pie plot shows that 59.2% of the total sales are shipped using standard class, which means most items ordered are shipped using standard class. The same-day ship mode has only 5.6%, which is the lowest.

## Segments
- The consumer segment accounts for 50.9% of the total sales, followed by the corporate segment with 30.3%, and lastly the home office segment with 18.8%.


```python
#convert date columns to datetime
store['Ship Date'] = pd.to_datetime(store['Ship Date'], format='%d/%m/%Y')
store['Order Date'] = pd.to_datetime(store['Order Date'], format='%d/%m/%Y')

#create three new columns using order date
store['Order Day'] = store['Order Date'].dt.day
store['Order Month'] = store['Order Date'].dt.month
store['Order Year'] = store['Order Date'].dt.year

#check the dataset
store.head(2)
```

```python
#line plot of order day, month and yea
plt.figure(figsize=(21, 5))

plt.subplot(1, 3, 1)
sns.set_style('dark')
sns.lineplot(x = 'Order Day', y = 'Sales',data = store, estimator = sum, errorbar=None)
plt.xlabel("Order Day", **{'weight':'bold', 'color':'#2a2e33'})
plt.ylabel("Total Sales", **{'weight':'bold', 'color':'#2a2e33'})
plt.xticks(rotation=90)


plt.subplot(1, 3,2)
sns.set_style('dark')
sns.lineplot(x = 'Order Month', y = 'Sales',data = store, estimator = sum, errorbar=None)
plt.xlabel("Order Month", **{'weight':'bold', 'color':'#2a2e33'})
plt.ylabel("Total Sales", **{'weight':'bold', 'color':'#2a2e33'})
plt.xticks(rotation=90)


plt.subplot(1, 3, 3)
sns.set_style('dark')
sns.lineplot(x = 'Order Year', y = 'Sales',data = store, estimator = sum, errorbar=None)
plt.xlabel("Order Year", **{'weight':'bold', 'color':'#2a2e33'})
plt.ylabel("Total Sales", **{'weight':'bold', 'color':'#2a2e33'})
plt.xticks(rotation=90, ticks=[2015, 2016, 2017, 2018])

plt.show()
```

* Day 17 exhibits the highest total sales, while day 29 experiences the lowest sales. This suggests potential fluctuations in sales performance throughout the month.

* Examining sales by month, November stands out with the highest total sales, indicating heightened purchasing activity during this period. Conversely, February registers the lowest total sales, reflecting potentially slower market conditions or seasonal factors.

* The total sales for 2015 show an initial increase, followed by a decline. However, sales rebound in 2016 and have continued to rise steadily since then, indicating overall growth and market resilience over the years.

```python
#plot the states with the highest and lowest sales
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
sns.set_style("darkgrid")
store.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar', color='#008b8b')
plt.title('TEN STATES WITH THE HIGHEST SALES')

plt.subplot(1,2,2)
sns.set_style("darkgrid")
store.groupby('State')['Sales'].sum().sort_values(ascending=False).tail(10).plot(kind='bar', color= '#fa8072')
plt.title('TEN STATES WITH THE LOWEST SALES')
```

California has the highest total sale, showcasing robust market performance and significant purchasing activity.

North Dakota has the lowest,  signaling comparatively subdued market activity in the region
